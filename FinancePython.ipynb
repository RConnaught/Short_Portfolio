## Moving Average

Before we continue, look at GM

### What Is a <!-- [Text](link) --> [Moving Average](https://www.investopedia.com/terms/m/movingaverage.asp) (MA)?
In statistics, a moving average is a calculation used to analyze data points by creating a series of averages of different subsets of the full data set. In finance, a moving average (MA) is a stock indicator that is commonly used in <!-- [Text](link) -->[technical analysis](https://www.investopedia.com/terms/t/technicalanalysis.asp). The reason for calculating the moving average of a stock is to help smooth out the price data by creating a constantly updated <!-- [Text](link) --> [average price](https://www.investopedia.com/terms/a/averageprice.asp).

By calculating the moving average, the impacts of random, short-term fluctuations on the price of a stock over a specified time frame are mitigated.

The previous figure is not a true represenation as it appears that Tesla dominanted, while GM and Ford's stock are relatively smooth and non-competitive by comparison.

```python
gm[['Date','Open']].plot(figsize=(15, 7))
```

There's a lot of noise going on. The data might not be as relable as we may think. This is where the **moving average** comes in. A method of making the data more reliable and robust against noise.

Create new column, moving average 50. The 50 represents the size of the window that will be smoothed over, 50 samples, 50 dates.

```python
gm['Open'].plot(figsize=(15, 7))
gm['MA50'] = gm['Open'].rolling(50).mean()
gm['MA50'].plot(label='MA50')
```

```python
gm['Open'].plot(figsize=(15, 7))
gm['MA50'] = gm['Open'].rolling(50).mean()
gm['MA100'] = gm['Open'].rolling(100).mean()
gm['MA50'].plot(label='MA50')
gm['MA100'].plot(label='MA100')
plt.legend()
```

#### Note:

When you're moving average window size is too high, you might miss out on some trends. You could apply auto regression that, or apply a combination of auto regression and moving average. Or work on the ACF (Auto Correlation Function). Or, use the EWMA (Exponentially Weighted Moving Average) among other things.

Techniques that stock analysts use when studying certain trends.

## Daily Percentage Change

### $r_t = \frac{p_t}{p_{t-1}} - 1$


R sub $t$ is the return at time $t$ and it's equal to the price at time $t$ divided by the price at time $t - 1$. The value represents a percentage gain or loss. The greater $r_t$ is the better the result and vice-versa. Not necessarily helpful in predicting the future value of the stock, it IS helpful in analyzing the volatillity of the stock.

For example, if your daily returns have a wide distributions the stock is more volatile

Percentage change is used for many purposes in finance, often to represent the price chnage of a stock over time, expressed as a percentage. The formula used to calculate percentage change is a simple mathematical concept.

- Calculate the percentage returns
- Plot them on histogram
- Base on the histogram, decide which stock is the most stable
    - Stable stocks are ideal, the less stable a stock is the higher the risk

```python
tesla['Returns'] = (tesla['Close'] / tesla['Close'].shift(1)) - 1
```

```python
ford['Returns'] = (ford['Close'] / ford['Close'].shift(1)) - 1
gm['Returns'] = (gm['Close'] / gm['Close'].shift(1)) - 1
```

## Cumulative Return

### What Is <!-- [Text](link) --> [Cumulative Return](https://www.investopedia.com/terms/c/cumulativereturn.asp)?
A cumulative return on an investment is the aggregate amount that the investment has gained or lost over time, independent of the amount of time involved. The cumulative return is expressed as a percentage, and it is the raw mathematical return of the following calculation:

$\frac{(\text{Current Price of Security}) - (\text{Original Price of Security})}{\text{Original Price of Security}}$

### The following formula we'll be using for this practice is this:

$i_t = (1+r_t)i_{t-1}$

This is a recursive formula. Multiple the cumulative return of day $t-1$ with a factor of plus $r_t$ (r sub t). What is $r_t$? The Daily Percentage Change. Accumulate the daily percentage return by plugging in $r_t$. By doing so, you'll get the following:

$i_t = (1+r_t)i_{t-1} = (1 + \frac{p_t}{p_{t-1}} - 1)i_{t-1} = \frac{p_t}{p_{t-1}} i_{t-1}$


```python
# Since the formula is recursive, make use of the .cumprod
tesla['Cumulative Returns'] = (1 + tesla['Returns']).cumprod()
```

`.cumprod()`: Returns the cumulative product of element along a given axis.

Read more <!-- [Text](link) --> [here](https://numpy.org/doc/stable/reference/generated/numpy.cumprod.html).

In this instance, the `.cumprod()` is $i_{t-1}$.

```python
tesla.head()
```

The first Cumulative Return should be $1 

```python
gm['Cumulative Returns'] = (1 + gm['Returns']).cumprod()
ford['Cumulative Returns'] = (1 + ford['Returns']).cumprod()
```

```python
tesla['Cumulative Returns'].plot(label='Tesla', figsize=(16, 8))
gm['Cumulative Returns'].plot(label='GM')
ford['Cumulative Returns'].plot(label='Ford')

plt.legend()


# Changes color of ticks, labels and axes to white
params = {"ytick.color" : "w",
          "xtick.color" : "w",
          "axes.labelcolor" : "w",
          "axes.edgecolor" : "w"}
plt.rcParams.update(params)
plt.title("Cumulative Return v. Time", color='white')
plt.show()
```
