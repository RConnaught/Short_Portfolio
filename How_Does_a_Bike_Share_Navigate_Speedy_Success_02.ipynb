```python
cyc_all_trips['ride_length'] = cyc_all_trips['ended_at'] - cyc_all_trips['started_at']
```

## Difference among riders

Visualize the difference among riders

```python
cyc_all_trips['member_casual'].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

cyc_all_trips.groupby('member_casual')['member_casual'].count().plot(kind='pie', autopct='%8.0f%%')

plt.title('Casual v. Membership riders')
```

```python
import plotly.express as px

#riders_df = cyc_all_trips['member_casual'].value_counts()
fig = px.pie(cyc_all_trips, values='member_casual', title='Casual v. Membership riders')
fig.show()
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Day', hue='member_casual', data=cyc_all_trips)
```

```python
sns.countplot(x='rideable_type', hue='member_casual', data=cyc_all_trips)
```

```python
sns.catplot(x='Month', hue='member_casual', col='rideable_type',
            data=cyc_all_trips, kind='count',
            height=5, aspect=.8)
```

```python
# Provide day of week with name
cyc_all_trips['day_of_week'] = cyc_all_trips['started_at'].dt.day_name()
```

```python
sns.set_style('whitegrid')
sns.countplot(x='day_of_week', hue='member_casual', data=cyc_all_trips)
```

## Share

Takeaway from analysis

- Although there are more riders with membership than casuals, the latter tend to use more bikes than those with membership during the weekend
- Casuals riders tend to ride more than twice as much on average than their counterparts with membership
- The classic bike seems to be vastly popular among riders, regardless of membership
- August is the most popular month for membership riders using classic bikes, while June is the popular month for casuals
- Among electric bike, between May and October seems to be the most popular months for membership riders, while July is the most popular with casuals
 - Although, the spike in popularity isn't anything to brag about

```python
member_by_days = cyc_all_trips.groupby(['member_casual', 'day_of_week'])
```

```python
avg_ride_length2 = member_by_days.ride_length.agg(['mean'])
```
